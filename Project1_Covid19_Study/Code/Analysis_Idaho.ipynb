```python
covid = pd.read_csv('/content/drive/MyDrive/United_States_COVID-19_Community_Levels_by_County.csv')
#drop null values
covid.dropna()
covid.head()
```

# Highest infected county in ID

```python
id_mask = covid['state'] == "Idaho"
idaho_covid = covid[id_mask]
idaho_covid
```

```python
# income.sort_values(by = 'Salary Paid', ascending = False, inplace = True)
# income = income.reset_index(drop=True)
# income

idaho_covid.sort_values(by = ['county', 'covid_cases_per_100k'],
                        ascending = [False,True], inplace = True)
idaho_covid = idaho_covid.reset_index(drop=True)
idaho_covid
```

```python

idaho_covid['date_updated'] = pd.to_datetime(idaho_covid['date_updated'])
```

```python
# Group by state and month, calculate median COVID cases per 100k population
county_covid_median_months = idaho_covid.groupby(['county', idaho_covid['date_updated'].dt.to_period('M')])['covid_cases_per_100k'].median()
```

```python
county_covid_median = county_covid_median_months.groupby('county').median()
county_covid_median.sort_values(ascending=False)
```

County with highest cases/100k

```python
print('The county with the highest median covid rate over 2022 was Elmore County with '
      + str(county_covid_median.iloc[0].round(2)) + ' cases per 100k people.')
```

#Lowest County in Idaho

```python
county_covid_median = county_covid_median_months.groupby('county').median()
county_covid_median.sort_values(ascending=True, inplace=True)
```

```python
print('The county with the lowest median covid rate over 2022 was Clark County with '
      + str(county_covid_median.iloc[0].round(2)) + ' cases per 100k people.')
```

# Hospital Admissions vs bed utilization overall

```python
# @title covid_hospital_admissions_per_100k vs covid_inpatient_bed_utilization

from matplotlib import pyplot as plt
x=idaho_covid['covid_hospital_admissions_per_100k']
y=idaho_covid['covid_inpatient_bed_utilization']

# Add scatterplot
plt.scatter(x, y, s=60, alpha=0.7, edgecolors="k")
b, a = np.polyfit(x, y, deg=1)
plt.xlabel('covid_hospital_admissions_per_100k')
plt.ylabel('covid_inpatient_bed_utilization')

plt.title('Hospital Admissions vs Inpatient Bed Use')
# Create sequence of 100 numbers from 0 to 100
xseq = np.linspace(0, 50, num=200)
# Plot regression line
plt.plot(xseq, a + b * xseq, color="k", lw=2.5);
```

```python
# 4. Progression Month to Month in Nevada
# Convert 'date_updated' column to datetime format
# Group by month and calculate aggregate metrics
monthly_stats_id = idaho_covid.groupby(idaho_covid['date_updated'].dt.to_period('M')).agg({
    'covid_inpatient_bed_utilization': 'mean'
})

# Plot metrics over time
plt.figure(figsize=(12, 6))
monthly_stats_id.plot(marker='o', ax=plt.gca())
plt.title('COVID Metrics Progression Month to Month in Idaho')
plt.xlabel('Month')
plt.ylabel('Metrics')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

```

##Elmore county

```python

plt.rcParams["figure.figsize"] = [10, 3.50]
plt.rcParams["figure.autolayout"] = True

elmore_cases.plot(xlabel='date_updated', ylabel='covid_cases_per_100k',x='date_updated', y='covid_cases_per_100k', alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)

plt.title("Elmore County", fontsize=16)
```
